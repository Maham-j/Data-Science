```python
#Part 3 SETS

page_a_users = {'user1', 'user2', 'user3', 'user4'}
page_b_users = {'user3', 'user4', 'user5', 'user6'}
page_c_users = {'user4', 'user7', 'user8'}


def find_common_users(set_a, set_b):
    return set_a.intersection(set_b)

def find_unique_users(set_a, set_c):
    return set_a.symmetric_difference(set_c)

def update_page_a(set_a, new_users):
    set_a.update(new_users)

def remove_users_from_page_b(set_b, users_to_remove):
    set_b.difference_update(users_to_remove)

page_a_users = {'user1', 'user2', 'user3', 'user4'}
page_b_users = {'user3', 'user4', 'user5', 'user6'}
page_c_users = {'user4', 'user7', 'user8'}

common_users = find_common_users(page_a_users, page_b_users)
print("Users who visited both Page A and Page B:", common_users)

unique_users = find_unique_users(page_a_users, page_c_users)
print("Users who visited either Page A or Page C, but not both:", unique_users)

new_users_for_a = {'user5', 'user9'}
update_page_a(page_a_users, new_users_for_a)
print("Updated Page A users:", page_a_users)

users_to_remove_from_b = ['user4', 'user5']
remove_users_from_page_b(page_b_users, users_to_remove_from_b)
print("Updated Page B users:", page_b_users)
```
